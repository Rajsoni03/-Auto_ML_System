# Auto Data Cleaning

Imorting libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import impute
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
import time

%matplotlib inline
```

## Upload Data File 
Accepting Formats are
- .csv
- .xlsx
- .json


```python
file_name = 'diabetes.csv'

file_format = file_name.split('.')[-1]

if file_format == 'csv':
    data = pd.read_csv(file_name)
elif file_format == 'xlsx':
    data = pd.read_excel(file_name)
elif file_format == 'json':
    data = pd.read_json(file_name)
else:
    print('File Does Not Supported.')
```

## Handile Missing Data

missing_data_type
- 0 (zero value)
- ?
- nan

```python
# Handile NaN/Null Data
tic = time.time()

strategy='mean'

columns = data.columns
imput   = impute.SimpleImputer(missing_values=np.NaN, strategy=strategy)
data    = list(imput.fit_transform(data))
data    = pd.DataFrame(data)

data.columns = columns

toc = time.time()
print(f'Cell Runtime {1000 * (toc-tic)} ms')
data
```

```python
data.info()
```

```python
# Handile Zero Data
tic = time.time()

missing_col = []
missing_data = data.isin([0])
for i in data.columns:
    if sorted(missing_data[i].unique())[-1]:
        missing_col.append(i)

toc = time.time()
print(f'Cell Runtime {1000 * (toc-tic)} ms')
missing_col
```

```python
print('Missing Data\n')
for i in missing_col:
    print(len(data[data[i]==0 ]), '\t:', i)
```

```python
data[missing_col].describe()
```

```python
# Handile NaN/Null Data
tic = time.time()

strategy='mean'
columns = data.columns

temp_data = data[missing_col_new]

imput   = impute.SimpleImputer(missing_values=0, strategy=strategy)
temp_data    = list(imput.fit_transform(temp_data))
temp_data    = pd.DataFrame(temp_data)

data[missing_col_new] = temp_data

data.columns = columns

toc = time.time()
print(f'Cell Runtime {1000 * (toc-tic)} ms')
data.head()
```

## Split Data into Features and Labels

- Get input from user fot Fratures (X)

```python
X = data[['Pregnancies',
          'Glucose', 
          'BloodPressure',
          'SkinThickness',
          'Insulin',
          'BMI', 
          'DiabetesPedigreeFunction', 
          'Age']]

Y = data['Outcome']
```

## Normalize the data

```python
columns = X.columns
X = pd.DataFrame(normalize(X, norm='max', axis=0))
X.columns = columns
X.head()
```

## Split Data to Train Test

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, train_size=.8, random_state=5)
```
